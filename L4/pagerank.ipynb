```python
# 使用networkX计算节点的pagerank
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 模拟计算过程(简化模型）
a = np.array([[0, 0, 0, 1/3, 0, 0],
             [1/4, 0, 0, 0, 1/2, 0],
             [0, 1, 0, 1/3, 1/2, 0],
             [1/4, 0, 0, 0, 0, 1],
             [1/4, 0, 1, 1/3, 0, 0],
             [1/4, 0, 0, 0, 0, 0]])
b = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
w = b
for i in range(100):
    w = np.dot(a, w)
    print(w) 
```

```python
# 模拟计算过程(随机模型）
a = np.array([[0, 0, 0, 1/3, 0, 0],
             [1/4, 0, 0, 0, 1/2, 0],
             [0, 1, 0, 1/3, 1/2, 0],
             [1/4, 0, 0, 0, 0, 1],
             [1/4, 0, 1, 1/3, 0, 0],
             [1/4, 0, 0, 0, 0, 0]])
b = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
w = b
d = 0.85
for i in range(100):
    w = (1-d)/6 + d*np.dot(a, w)
    print(w) 
```

```python
# 创建有向图
G = nx.DiGraph()
# 设置有向图的边集合
edges = [('A', 'B'), ('A', 'F'), ('A', 'E'), ('A', 'D'), ('B', 'C'), ('C', 'E'), ('D', 'A'), ('D', 'C'), ('D', 'E'),
         ('E', 'B'), ('E', 'C'), ('F', 'D')]
# 在有向图G中添加边集合
for edge in edges:
    G.add_edge(edge[0], edge[1])
```

```python
# 有向图可视化
layout = nx.circular_layout(G)
nx.draw(G, pos=layout, with_labels=True, hold=False)
plt.show()
```

```python
# 计算简化模型的PR值
pr = nx.pagerank(G, alpha=1)
print("简化模型的PR值：", pr)

# 计算随机模型的PR值
pr = nx.pagerank(G, alpha=0.8)
print("随机模型的PR值：", pr)
```
